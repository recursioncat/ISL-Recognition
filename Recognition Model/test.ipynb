```python
from tensorflow.keras.preprocessing import image
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
class_names = ['1',
 '2',
 '3',
 '4',
 '5',
 '6',
 '7',
 '8',
 '9',
 'a',
 'b',
 'c',
 'd',
 'e',
 'f',
 'g',
 'h',
 'i',
 'j',
 'k',
 'l',
 'm',
 'n',
 'o',
 'p',
 'q',
 'r',
 's',
 't',
 'u',
 'v',
 'w',
 'x',
 'y',
 'z']
```

```python
model = tf.keras.models.load_model('gesturesV4.keras')
```

```python
img_path  = '7.jpg'
img = image.load_img(img_path, target_size=(64,64))  # Resize to match the input size expected by your model

plt.imshow(img)
```

```python
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
print(img_array)
```

```python
predictions = model.predict(img_array)

top_5_indices = np.argsort(predictions[0])[-5:][::-1]
print("Top 5 predictions with percentages:")
for i in top_5_indices:
    print(f"{class_names[i]}: {predictions[0][i] * 100:.2f}%")
```
